# Выявление зависимости вакцинации от количества твитов о вакцинации

Используя тест Гренджера на причинность, необходимо определить, является ли
процесс А (динамика вакцинации от ковида) причиной процесса Б (твиты про ковид),и наобарот.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from statsmodels.tsa.stattools import grangercausalitytests
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss
```

## Подготовка датасетов

Датасеты состоят из большоего количества стран. Влияние твиттера на вакцинацию в некоторых страннах незначительно, поэтому возьмем топ 20 стран по частоте использования твиттера.

### Подготовим датасет вакцинации

```python
df_vac = pd.read_csv('vaccinations.csv')
df_vac
```

```python
type(df_vac.date[0])
```

Тип даты - str, переведем его в datetime

```python
df_vac.date = pd.to_datetime(df_vac.date)
```

Будем смотреть по колонкам date и daily_vaccinations. Уберем все строки с NaN в этих колонках

```python
df_vac = df_vac.dropna(subset=['daily_vaccinations'])
```

```python
df_vac.location.unique()
```

 Напишем функцию для поиска нужных стран, она будет полезнее на следующем датасете твитов

```python
def data_frame_from_countries(df, column: str):
    keywords = {
        'usa', 'america', 'japan', 'india', 'brazil', 'indonesia',
        'england', 'britain', 'uk', 'turkey', 'mexico', 'thailand',
        'philippines', 'france', 'spain', 'canada', 'ca','germany',
        'argentina', 'egypt', 'malaysia', 'colombia'
    }
    double_word_key = {
        'united', 'states', 'kingdom', 'south',
        'korea', 'saudi', 'arabia'
    }
    new_df = pd.DataFrame()
    for i, val in enumerate(df[column]):
        words = re.findall(r'\w+', val)
        words_low = [x.lower() for x in words]
        if set(words_low) & keywords or \
        len(set(words_low) & double_word_key) == 2:
            new_df = new_df.append(df.iloc[[i]])
    return new_df
```

```python
res_df_vac = data_frame_from_countries(df_vac, 'location')
```

```python
res_df_vac.location.unique()
```

```python
res_df_vac.location.unique().shape
```

Получилось на 3 страны больше. Добавились такие названия как South America, North America и England. Удалим их вручную (так как они включают в себя страны, которые уже есть)

```python
res_df_vac = res_df_vac[
    (res_df_vac.location != 'South America') & \
    (res_df_vac.location != 'North America') & \
    (res_df_vac.location != 'England')
]
```

```python
res_df_vac.location.unique().shape
```

Теперь сделаем датасет с суммой вакцинированных за день для всех стран, в которых популярен твиттер

```python
res_df_vac.loc[res_df_vac.duplicated(subset=['date'])]
```

```python
dates = res_df_vac.date.unique()
date_list = []
daily_vaccinations_list = []
for val in dates:
    vac_summ = res_df_vac[res_df_vac.date == val].daily_vaccinations.sum()
    date_list.append(val)
    daily_vaccinations_list.append(vac_summ)
    
df_all_vac = pd.DataFrame(
    {'date':date_list,
     'daily_vaccinations':daily_vaccinations_list}
)
```

```python
df_all_vac = df_all_vac.sort_values(by = 'date')
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df_all_vac.date, df_all_vac.daily_vaccinations)
```

## Подготовим датасет твитов

```python
df_tweets = pd.read_csv('vaccination_all_tweets.csv')
df_tweets
```

```python
type(df_tweets.user_created[0])
```

```python
df_tweets.user_created = pd.to_datetime(
    df_tweets.user_created, format="%Y-%m-%d"
)
df_tweets.user_created = pd.to_datetime(df_tweets.user_created.dt.date)
```

В user_location присутствуют NaN, уберем их

```python
df_tweets = df_tweets.dropna(subset=['user_location'])
```

История короновируса из первого датасета началась со следующей даты

```python
firts_date = min(df_all_vac.date)
firts_date
```

Можно отбросить все твиты до 2020 12 08

```python
df_tweets = df_tweets[df_tweets.user_created >= firts_date]
df_tweets
```

Применем функцию, которую мы сделали ранее для поиска стран для колонки user_location

```python
df_res_tweet = data_frame_from_countries(df_tweets, 'user_location')
```

```python
df_res_tweet
```

```python
df_res_tweet.user_created.unique().shape
```

Теперь сложим все твиты за один месяц

```python
dates = df_res_tweet.user_created.unique()
date_list = []
tweet_count_list = []
for val in dates:
    count = 0
    count += df_res_tweet[df_res_tweet.user_created == val].shape[0]
    date_list.append(val)
    tweet_count_list.append(count)
    
df_all_tweet = pd.DataFrame(
    {'date':date_list, 'tweet_number':tweet_count_list}
)
```

```python
df_all_tweet = df_all_tweet.sort_values(by='date')
```

Построим график по количеству твитов за месяц

```python
plt.figure(figsize=(10, 5))
plt.plot(df_all_tweet.date, df_all_tweet.tweet_number)
```

## Соединим датасеты

```python
merged_df = pd.merge(df_all_vac, df_all_tweet, how='inner', on ='date' )
merged_df
```

## Применим тест Гренджера

Нулевая гипотеза: количество твитов о вакцинаяии за день НЕ является причиной количества вакцинаций за день

```python
grangercausalitytests(
    merged_df[['daily_vaccinations', 'tweet_number']], maxlag=[3]
)
```

p-значение = 0.0117, меньше 0.05, следовательно мы можем отклонить нулевую гипотезу теста и сделать вывод, что количество твитов о вакцинации полезно для прогнозирования будущего количества дневных вакцинаций

Проверим обратную гипотезу

Нулевая гипотеза: количество вакцинаций за день НЕ является причиной количества твитов о вакцинации за день

```python
grangercausalitytests(
    merged_df[['tweet_number', 'daily_vaccinations']], maxlag=[3]
)
```

p-значение = 0.0671, больше 0.05, следовательно мы НЕ можем отклонить нулевую гипотезу теста и можем сделать вывод, что количество вакцинаций за день не является причиной будущего количества твитов за день

## Определить, являются ли данные процессы стационарными

### Воспользуемся тестом Дики-Фуллера

При помощи этого теста проверяют значение коэффициента автореграсии, если он равен 1, то процесс не стационарен, если он больше 1, то процесс стационарный

Нулевая гипотеза: процесс является НЕ стационарным

```python
test_stat, pvalue, n_lags, _, _, _  = adfuller(
    merged_df['tweet_number'], autolag='AIC'
)
print(f'test_stat = {test_stat}\npvalue = {pvalue}\nn_lags = {n_lags}')
```

Т.к. p-значение меньше 0.05, то нулевую гипотезу можно опровергнуть, значит процесс количества твитов о вакцинации в день стационарный

```python
test_stat, pvalue, n_lags, _, _, _  = adfuller(
    merged_df['daily_vaccinations'], autolag='AIC'
)
print(f'test_stat = {test_stat}\npvalue = {pvalue}\nn_lags = {n_lags}')
```

Т.к. p-значение больше 0.05, то нулевую гипотезу нельзя опровергнуть, значит процесс количества вакцинации в день НЕ стационарный

# Выводы

1) Количество твитов о вакцинации является причиной количества вакцинаций в странах, где твитер имеет высокую поплуярность

2) Процесс количества твитов в день является стационарным временным рядом, когда количество вакцинаций в день не стационарным
